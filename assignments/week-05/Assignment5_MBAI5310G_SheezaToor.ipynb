```python
df_clean = df.copy()

print("Original dataset shape:", df_clean.shape)
```

```python
df_clean.columns = df_clean.columns.str.strip()
df_clean.columns
```

```python
duplicate_count = df_clean.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)
```

```python
df_clean = df_clean.drop_duplicates()
print("Dataset shape after removing duplicates:", df_clean.shape)
```

```python
numerical_columns = [
    "Customer_Tenure_Months", 
    "Avg_Monthly_Card_Spend", 
    "Travel_Spend_Share_Pct", 
    "Dining_Spend_Share_Pct", 
    "Digital_App_Logins_30D", 
    "Reward_Points_Balance", 
    "Prior_Offer_Emails_Opened_6M", 
    "Credit_Utilization_Pct", 
    "Late_Payments_12M",
    "Customer_Satisfaction_Score"
]

df_clean["Avg_Monthly_Card_Spend"] = (
    df_clean["Avg_Monthly_Card_Spend"]
    .replace(r"[\$,]", "", regex=True)
    .astype(float)
)

for col in numerical_columns:
    df_clean[col] = df_clean[col].fillna(df_clean[col].median())

df_clean.isnull().sum()
```

```python
X = df_clean[numerical_columns]

X.head()
```

```python
X.isnull().sum()
```

```python
#Scaling Features 

from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns=numerical_columns)
X_scaled_df.head()
```

```python
#Mean and Standard Deviation after Scaling

print("Mean after scaling:")
print(X_scaled_df.mean())
print("\nStandard deviation after scaling:") 
print(X_scaled_df.std())
```

```python
#Choose Number of Clusters 

import pandas as pd

k_options = pd.DataFrame({
    "K Value": [2, 3, 4, 5, 6],
    "Meaning": [
        "Customers are divided into 2 broad groups",
        "Customers are divided into 3 main groups",
        "Customers are divided into 4 more detailed groups",
        "Customers are divided into 5 smaller groups",
        "Customers are divided into 6 very specific groups"
] })
k_options

print("We need to choose a suitable number of clusters.")
print("The number of clusters is called K.")
print("Next, we will use the Elbow Method to compare different K values.")
```

```python
#Elbow Method for Finding Best K

from sklearn.cluster import KMeans 
import matplotlib.pyplot as plt

inertia = []

K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) 
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method graph

plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Finding the Best K")
plt.show()
```

```python
#Displaying Intertia Values for Each K

elbow_results = pd.DataFrame({
    "K": list(K_range),
    "Inertia": inertia
})
elbow_results
```

```python
#Training K-Means Model

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10) 

cluster_labels = kmeans.fit_predict(X_scaled)

cluster_labels[:10]
```

```python
#How many customers in each cluster?

import pandas as pd 

pd.Series(cluster_labels).value_counts().sort_index()
```

```python
#Adding Cluster labels to Dataset

df_clustered = df_clean.copy()

df_clustered["Cluster"] = cluster_labels

df_clustered.head()
```

```python
#How many customers in each cluster: 

df_clustered["Cluster"].value_counts().sort_index()
```

```python
numerical_columns = [
    "Customer_Tenure_Months",
    "Avg_Monthly_Card_Spend",
    "Travel_Spend_Share_Pct",
    "Dining_Spend_Share_Pct",
    "Digital_App_Logins_30D",
    "Reward_Points_Balance",
    "Prior_Offer_Emails_Opened_6M",
    "Credit_Utilization_Pct",
    "Late_Payments_12M",
    "Customer_Satisfaction_Score"
]

cluster_summary = df_clustered.groupby("Cluster")[numerical_columns].mean()
print(cluster_summary)

cluster_summary_rounded = cluster_summary.round(2)
cluster_summary_rounded
```

```python
#Number of customers in each cluster 

cluster_counts = df_clustered["Cluster"].value_counts().sort_index()
cluster_counts
```

```python
#Summary table with customer counts and average feature values 

cluster_summary_final = cluster_summary_rounded.copy()
cluster_summary_final["Number_of_Customers"] = cluster_counts
cluster_summary_final
```

```python
#Interpeting Customer/User Segments 

cluster_summary_final
```

```python
#Identifying clusters based on customer behaviour 

summary = df_clustered.groupby("Cluster").mean(numeric_only=True)

# Highest spenders
high_spender_cluster = summary["Avg_Monthly_Card_Spend"].idxmax()

# Remaining clusters
remaining_clusters = summary.drop(index=high_spender_cluster)

# Most engaged among remaining
premium_cluster = remaining_clusters["Digital_App_Logins_30D"].idxmax()

# Last cluster = low engagement
low_engagement_cluster = summary.drop(
    index=[high_spender_cluster, premium_cluster]
).index[0]

segment_names = {
    high_spender_cluster: "High Value Spenders",
    premium_cluster: "Highly Engaged Customers",
    low_engagement_cluster: "Low Engagement / At-Risk Customers"
}

df_clustered["Segment_Name"] = df_clustered["Cluster"].map(segment_names)
df_clustered.head()
```

```python
#Interpretation table 

segment_interpretation = pd.DataFrame({
    "Cluster": list(segment_names.keys()),
    "Segment Name": list(segment_names.values()),
    "Business Interpretation": [
        "Customers with high monthly card spend and strong purchasing activity across categories (travel, dining).",
        "Customers with high digital engagement and strong overall activity, likely responsive to offers and rewards programs.",
        "Customers with low spending, low engagement, and lower product interaction, potentially at risk of churn."
    ],
    "Possible Business Strategy": [
        "Offer loyalty rewards, cashback incentives, and exclusive premium promotions to retain high-value spenders.",
        "Promote upsell opportunities, personalized offers, and reward-based engagement to increase value per customer.",
        "Run reactivation campaigns, targeted discounts, and engagement emails to improve activity and retention."
    ]
})

segment_interpretation
```

```python
#Visualization using two features

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(
    df_clustered["Avg_Monthly_Card_Spend"],
    df_clustered["Digital_App_Logins_30D"],
    c=df_clustered["Cluster"],
    alpha=0.7
)

plt.xlabel("Average Monthly Card Spend")
plt.ylabel("Digital App Logins (30D)")
plt.title("Customer Clusters: Spending vs Engagement")
plt.colorbar(label="Cluster")
plt.show()
```

```python
#Visualization using segment names

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

for segment in df_clustered["Segment_Name"].unique():
    segment_data = df_clustered[df_clustered["Segment_Name"] == segment]

    plt.scatter(
        segment_data["Avg_Monthly_Card_Spend"],
        segment_data["Digital_App_Logins_30D"],
        label=segment,
        alpha=0.7
    )

plt.xlabel("Average Monthly Card Spend")
plt.ylabel("Digital App Logins (30D)")
plt.title("Customer Segments Using Spending vs Engagement")
plt.legend()
plt.show()
```
